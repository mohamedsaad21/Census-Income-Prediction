## Data Description
| Feature | Type | Description |
|---------|------|-------------|
| age | Numeric | Age of the person in years |
| workclass | Categorical Nominal | Type of employer |
| fnlwgt | Numeric | Census weight for the record |
| education | Categorical Nominal | Highest education level |
| education-num | Categorical Ordinal | Numeric ranking of education level |
| marital-status | Categorical Nominal | Marital status of the person |
| occupation | Categorical Nominal | Job role or profession |
| relationship | Categorical Nominal | Relationship within the family |
| race | Categorical Nominal | Race category |
| sex | Categorical Binary | Male or Female |
| capital-gain | Numeric | Capital gains from investments |
| capital-loss | Numeric | Capital losses from investments |
| hours-per-week | Numeric | Number of working hours per week |
| native-country | Categorical Nominal | Country of origin |
| income | Target Binary | <=50K low income and >50K high income |

### 3.2: Label Encoding

```python
import pickle
from sklearn.preprocessing import LabelEncoder
```

```python
# Dictionary to hold one encoder per column
encoders_dict = {} 

for col in dataset.columns:
    if dataset[col].dtypes == 'object':
        le = LabelEncoder()
        dataset[col] = le.fit_transform(dataset[col])
        # Save this specific column's encoder into the dictionary
        encoders_dict[col] = le

# Save the DICTIONARY, not just the last encoder
with open("encoders.pkl", "wb") as f:
    pickle.dump(encoders_dict, f)
```

### 3.3: Feature Selection

```python
X = dataset.drop('income', axis=1)
Y = dataset['income']
```

### 3.4: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Define X and Y (Same as you had)
X = dataset.drop('income', axis=1)
Y = dataset['income']

# 2. Scale everything together
scaler = StandardScaler()
X_scaled_array = scaler.fit_transform(X)

# 3. CRITICAL STEP: Convert back to DataFrame
# This ensures X keeps its column names and "looks" the same to the rest of your code
X = pd.DataFrame(X_scaled_array, columns=X.columns)

# 4. Save the scaler
import pickle
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

# Now you can check it
print(X.head())
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_scores = cross_val_score(
    model,
    X,       
    Y,
    cv=skf,
    scoring='accuracy'
)
```

### 3.5: Fixing imbalanced dataset using Oversampling

```python
round(Y.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
```

```python
ros.fit(X, Y)
```

```python
X_resampled, Y_resampled = ros.fit_resample(X, Y)
```

```python
round(Y_resampled.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

### 3.6: Model Training and Evaluation with Train/Test Split and Cross-Validation


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# 1) Train / Test Split
X_train, X_test, Y_train, Y_test = train_test_split(
    X_resampled, Y_resampled, test_size=0.2, random_state=42, stratify=Y_resampled)


# 2) Train the model
model.fit(X_train, Y_train)

# 3) Predict on test set
Y_test_pred = model.predict(X_test)

# 4) Calculate accuracy
test_accuracy = accuracy_score(Y_test, Y_test_pred)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```

```python
print(f"Train/Test Accuracy: {test_accuracy:.4f}")
print(f"CV Mean Accuracy: {cv_scores.mean():.4f}")
print("Cross-Validation Accuracy Scores:", cv_scores)
print(f"CV Mean Accuracy: {cv_scores.mean():.4f}")
print(f"CV Std Deviation: ±{cv_scores.std():.4f}")
```

## Step 4: Data Modelling

# Basic Models

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import pandas as pd

Basic_models = {
    "Logistic Regression": LogisticRegression(),
    "Naive Bayes": GaussianNB(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}
```

```python
results = []

print("===== Basic Models Evaluation =====\n")

for name, model in Basic_models.items():
    model.fit(X_train, Y_train)

    y_pred = model.predict(X_test)

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(Y_test, y_pred),
        "Precision": precision_score(Y_test, y_pred),
        "Recall": recall_score(Y_test, y_pred),
        "F1-score": f1_score(Y_test, y_pred)
    })

    print(f"{name} Accuracy: {accuracy_score(Y_test, y_pred):.3f}")
    print(classification_report(Y_test, y_pred))
    print("*******************************************\n")

results_df = pd.DataFrame(results)
results_df
```

## Step 6: Hyperparameter Tuning

```python
# Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV

param_grids = {
    "Logistic Regression": {
        "C": [0.01, 0.1, 1, 10],
        "solver": ["lbfgs"],
        "max_iter": [200, 500, 1000]
    },
    "Naive Bayes": {},  # No params
    "KNN": {
        "n_neighbors": [3, 5, 7, 9],
        "weights": ["uniform", "distance"]
    },
    "Decision Tree": {
        "max_depth": [3, 5, 10, None],
        "min_samples_split": [2, 5, 10]
    },
    "Random Forest": {
        "n_estimators": [50, 100, 200],
        "max_depth": [3, 5, 10, None],
        "min_samples_split": [2, 5, 10]
    }
}
```

```python
tuned_models = {}
tuned_results = []

print("===== Hyperparameter Tuning =====\n")

for name, model in Basic_models.items():

    print(f"Running GridSearch for {name}...\n")

    # If model has a param grid
    if param_grids.get(name):
        grid = GridSearchCV(
            estimator=model,
            param_grid=param_grids[name],
            scoring="f1",
            cv=5,
            n_jobs=-1
        )
        grid.fit(X_train, Y_train)
        best_model = grid.best_estimator_
        best_params = grid.best_params_

    else:
        # Models with no tuning (like Naive Bayes)
        model.fit(X_train, Y_train)
        best_model = model
        best_params = "No tuning needed"

    tuned_models[name] = best_model

    y_pred = best_model.predict(X_test)

    tuned_results.append({
        "Model": name + " (Tuned)",
        "Accuracy": accuracy_score(Y_test, y_pred),
        "Precision": precision_score(Y_test, y_pred),
        "Recall": recall_score(Y_test, y_pred),
        "F1-score": f1_score(Y_test, y_pred),
        "Best Params": best_params
    })

    print(f"{name} Tuned Accuracy: {accuracy_score(Y_test, y_pred):.3f}")
    print(classification_report(Y_test, y_pred))
    print(f"Best Params: {best_params}")
    print("*******************************************\n")

tuned_results_df = pd.DataFrame(tuned_results)
tuned_results_df
```

```python
all_models = pd.concat([results_df , tuned_results_df] , axis =0).sort_values('Model').reset_index()
all_models
```
